```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import json
from pathlib import Path
from PIL import Image

import sys
sys.path.insert(0, "..")

from src.environment import GridWorldWithWallsEnv, GridWorldEnv
```

```python
save_dir = "epsilon_decay"
save_dir = Path("../runs") / save_dir

with open(save_dir / "eval_results.json") as f:
    results = json.load(f)

episode_lengths = np.array(results["episode_lengths"])
```

```python
x = episode_lengths
x = x[x < 100]
```

```python
# Episode length histogram
plt.figure(figsize=(10, 6))
sns.histplot(x, bins=range(1, max(x) + 2), edgecolor='black')
plt.title('Episode Length Distribution')
plt.xlabel('Episode Length')
plt.ylabel('Frequency')
plt.show()
```
